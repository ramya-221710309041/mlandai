```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Boston Dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
## To find the correaltion we use corr()
boston_df.corr()
```

```python
#### Visualizing the correlation
plt.figure(figsize=(14, 14))
sns.heatmap(boston_df.corr(), annot=True, cmap='YlGnBu')
```

```python
#correlation-->0.7
correlation = boston_df.corr()

cor = correlation[correlation>0.7]
plt.figure(figsize=(13, 13))
sns.heatmap(cor, cmap='Greens', annot=True)
```

```python
#checaking the correlation between 0.7 & 0.9
cor = correlation[(correlation>0.7) & (correlation<0.9)]
cor
```

```python
#All the columns-- as input
##Drop the columns which are highly correlated and biuld a MLR; you need to compare the performance
```

```python
##Split the data into traning and testing; Build the model on training data and test the model on testing data
```

```python
#Split the data into Input and Output
X = boston_df.drop(['MEDV'], axis=1) #Input columns
y = boston_df.MEDV #boston_df['MEDV']
```

```python
#Divide the X and y into train and test
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X, y, test_size= 0.25,
                                                   random_state=0)
```

```python
#Build the model on traning data --> X_train and y_train
#sklearn library: import, instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
#Intercept and the coefficient values
print(lm.intercept_)
lm.coef_
```

```python
#create a dataframe for coefficients
coefficients = pd.DataFrame([X_train.columns, lm.coef_]).T
coefficients
```

```python
#y = b0+b1X1+b2X2+.....b13X13
#y = 36.93325545711978 +(-0.117735)()
```

```python
#Checking the model prediction on traning data
y_train_pred = lm.predict(X_train)
y_train_pred
```

```python
#we need to compare the actual values(y_train) and the predicted 
                                                       #values(y_train_pred)
from sklearn.metrics import r2_score
print('R^2:', r2_score(y_train,y_train_pred))
print('Adjusted R^2:', 1- (1-r2_score(y_train, y_train_pred))*(len(X_train)-1)/
                                (len(X_train)-X_train.shape[1]-1))
```

```python
#regplot(y_train--> actual values and y_train_pred-->pred)
sns.regplot(y_train, y_train_pred)
```

```python
plt.scatter(y_train, y_train_pred)
plt.xlabel('Price')
plt.ylabel('Model Predicted Price')
```

```python
#plot for residuals(Errors)
#y_train-y_train_pred==Errors
plt.scatter(y_train_pred, y_train-y_train_pred)
plt.xlabel("Predicted Values")
plt.ylabel("residuals")
```

```python
#Test the model on testing data
y_test_pred = lm.predict(X_test) #test data--> unseen data
y_test_pred
```

```python
#we need to compare the actual values(y_train) and the predicted 
                                                       #values(y_train_pred)
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print('R^2:', r2_score(y_train,y_train_pred))
print('Adjusted R^2:', 1- (1-r2_score(y_test, y_test_pred))*(len(X_test)-1)/
                                (len(X_test)-X_test.shape[1]-1))

print('MAE:', mean_absolute_error(y_test, y_test_pred))

print('MSE:', mean_squared_error(y_test, y_test_pred))

print('RMSE', np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

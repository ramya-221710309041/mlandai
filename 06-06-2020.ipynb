```python
#renaming the column header
import pandas as pd
import numpy as np
s1 = pd.Series([9,11,1,3,4,5,6,2,9])
s2 = pd.Series([1.1,3.5,4.7,5.8,2.9,9.3,1.4,2.9,6.4])
s3 = pd.Series(['a','b','c','d','e','f','g','h','i','j'])
data = {"first":s1,"second":s2,"third":s3}  
df = pd.DataFrame(data)
df
```

```python
df.rename(columns = {"second":"column2","third":"column3"},inplace=True)
```

```python
df
```

### Changing the row index:
1. set_index()
2. reset_index()


```python
df.set_index('column3')
df
```

```python
df=df.reset_index()
df
```

```python
label = list("abcdefghi")
label
```

```python
df = df.set_index([pd.Index(label)])
df
```

### Removing columns:

```python
#Single column removing
df.drop(['first'],axis=1)
```

```python
df.drop(columns = ["first","second"])
```

```python
df.drop(df.columns[[0,2]],axis=1)
```

### Removing Rows:

```python
#single row
df.drop([7])
```

```python
#remove mutliple rows
df.drop([7,8])
```
